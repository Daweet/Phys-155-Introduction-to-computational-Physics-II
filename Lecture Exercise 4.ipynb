# Coupled Differential Equation

Harmonic Oscillator with damping b

$\frac{dv}{dt}=-k*x-b*v$ also considering $\frac{dx}{dt}=v$

Simple Harmonic Oscillator with Damping

Using the vector $Y(x,v)=\begin{pmatrix}x  \\ v \end{pmatrix}$ we can write the coupled differential equation in a matrix can be written as $\frac{dY}{dt}=M*Y$ 

$M(x,v)=\begin{pmatrix}0  & 1\\ -k & -b\end{pmatrix}$ we should get the same answer when you check.

$\frac{d}{dt}\begin{pmatrix}x  \\ v \end{pmatrix}=\begin{pmatrix}0  & 1\\ -k & -b\end{pmatrix}*\begin{pmatrix}x  \\ v \end{pmatrix}$ 


```python
# import all your libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def sho_equations(y,t,k,b):
    x,v = y
    dxdt = v
    dvdt = -k*x-b*v
    return [dxdt,dvdt]

y0 = [1.0, 0.0]# Initial condition
k =1.0
b=0.0
t = np.linspace(0,30,1000)

soln = odeint(sho_equations,y0,t,args=(k,b))

x_val = soln[:,0]
v_val = soln[:,1]

plt.plot(t,x_val, label='Postion x')
plt.plot(t,v_val, label='Velocity')
plt.xlabel('Time')
plt.ylabel('Position/Velocity')
plt.legend()
plt.title('Simple Harmonic Oscillator, zero damping')
plt.show
```
