# Dissipating System
### in such case we need to consider density matrix instead of state

$\rho = |\psi(t)\rangle\langle\psi(t)|$

solving $\frac{d\rho}{dt}=L\rho$, where has dissipative terms like relaxation

$\rho(t)= e^{Lt}\rho(0)$

```python
#import your libraries first
import numpy as np
import matplotlib.pyplot as plt #Library for plotting
%matplotlib inline
#OBE = Optical Bloch Equations
import OBE_Tools as OBE
import time
```

```python
Omegas=[1]
Deltas=[0]
Gammas=[Omegas[0]/10]
L=OBE.time_dep_matrix(Omegas,Deltas,Gammas)

times = np.linspace(0,8*np.pi,500)
rho_0 = np.transpose(np.asarray([1,1,1,0]))
rho_t = np.zeros((len(times),4))
pops = np.zeros((len(times),2))
cohrs = np.zeros((len(times),2))

for i,t in enumerate(times):
    sol = OBE.time_evolve(L, t,rho_0)# Evolution
    rho_00,rho_11 = np.real(sol[0]), np.real(sol[3])
    rho_01,rho_10 = np.real(sol[1]), np.real(sol[2])
    norm = rho_00 + rho_11
    popn = [rho_00,rho_11]
    #cohr = [rho_01,rho_10]
    pops [i,:]=popn/norm
    #cohr [i,:]=cohrs/norm
    
plt.figure()
plt.plot(times/(2*np.pi), pops[:,0], label =r'$|\psi_0\rangle$')
plt.plot(times/(2*np.pi), pops[:,1], label =r'$|\psi_1\rangle$')
plt.legend()
plt.xlabel('Time ($2\pi/\Omega$)')
plt.ylabel('Population of states 0 and 1')
```
