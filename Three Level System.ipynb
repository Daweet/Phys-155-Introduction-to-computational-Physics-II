```python
import numpy as np
import matplotlib.pyplot as plt
import OBE_Tools as OBE
import time
```

```python
Omegas = [1,5] #2pi MHz(Pulse area)
Deltas = [0,0] #2pi MHz  in reduced units
Gammas = [0.5, 0.1]
```

```python
Delta_01 = np.linspace(-10,10,200)
probe_abs = np.zeros((2,len(Delta_01)))

for i, p in enumerate(Delta_01):
    Deltas[0] = p
    solu = OBE.steady_state_soln(Omegas, Deltas, Gammas)
    probe_abs[0,i]= np.imag(solu[1])
    solu = OBE.steady_state_soln([1,0], Deltas, Gammas)
    probe_abs[1,i]= np.imag(solu[1])
    
    
```

```python
plt.figure()
plt.plot(Delta_01, probe_abs[1,:], label=r'$\Omega_{12} = 0$MHz')#Blue
plt.plot(Delta_01, probe_abs[0,:], label=r'$\Omega_{12} = 5$MHz')#Orange
#Plt.xlabel(detuning), ylabel(imaginary part of rho_01)
#legend
plt.show()
```
